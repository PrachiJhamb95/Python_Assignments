```python
import pandas as pd
import re
import datetime as dt
pd.set_option('display.max_colwidth', None)
```

```python
#read the TS in Harry Potter 1.csv (note, this is a semicolon separated dataset)
#this dataset was retrieved from https://www.kaggle.com/datasets/gulsahdemiryurek/harry-potter-dataset
df = pd.read_csv("data/Harry Potter 1.csv", sep = ";")
display(df)
```

```python
#get all sentences that end in a question mark spoken by Ron (hint, use the str.contains method of a series)
df_Ron = df[df["Character"]== "Ron"]
df_Ron[df_Ron["Sentence"].str.contains(r"\.*.*\?$")]
```

```python
#get all of the sentences that mention both Harry and Ron
df_Ron_Harry = df[df["Sentence"].str.contains(r"(?=.*Harry)(?=.*Ron)")]
df_Ron_Harry
```

```python
#get all of the sentences that start with a number
df_num = df[df["Sentence"].str.contains(r"(^[0-9])")]
df_num
```

```python
#get all sentences that mention a professor at the end of a sentence excluding punctuation, and extract just name of the professor mentioned (hint, concatenate the result of the extract operation with the original dataframe along axis=1) and then dropna() to eliminate the ones that had no mention
df_prof = df[df["Sentence"].str.contains(r"(?<=Professor ).\w+\.|(?<=Professor ).\w+\?")] 
df_prof1 = df_prof.Sentence.str.extract(r"((?<=Professor ).\w+)",flags=re.IGNORECASE)  
df_Professor = pd.concat([df_prof1,df_prof],axis=1) 
df_Professor.rename(columns={0:"Mentioned"},inplace=True) 
display(df_Professor)
```

```python
#extract quotes contained within any sentence
df_quote = df[df["Sentence"].str.contains(r'"([^"]*)"')]
df_quoted = df_quote.Sentence.str.extract(r'"([^"]*)"',flags=re.IGNORECASE)
df_concat = pd.concat([df_quoted,df_quote],axis=1)
df_concat.rename(columns={0:"Quoted"},inplace=True)
display(df_concat)
```

```python
#extract the first hyphenated word in each sentence, where the first word is at least 2 characters
df_hyphenated = df.loc[df.Sentence.str.contains(r"(?=\S{2,}[-])([a-zA-Z-]+)"),:]
df_hyph = df_hyphenated.Sentence.str.extract(r"(?=\S{2,}[-])([a-zA-Z-]+)",flags=re.IGNORECASE)
df_output = pd.concat([df_hyph,df_hyphenated],axis=1)
df_output.rename(columns={0:"Hyphenated"},inplace=True)
display(df_output)
```
